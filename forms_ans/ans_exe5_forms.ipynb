```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dados_maquinas = pd.DataFrame({
    'maquina':['m1','m2','m3','m4','m5','m6','m7','m8','m9','m10'],
    'temperatura': [70.2,65.1,75.5,80.3,68.7,72.9,78.6,66.4,73.1,69.5],
    'vibracao' : [12.5,8.2,15.6,10.2,11.8,14.3,9.8,8.9,13.7,12.1],
    'corrente' : [4.7,3.9,5.1,4.5,4.2,5.3,4.8,4.0,5.0,4.3],
})
```

```python
dados_maquina = pd.get_dummies(dados_maquinas, columns=['maquina'])
```

```python
dados_maquina
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(dados_maquina)
```

```python
sse = kmeans.inertia_
print(sse)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```
