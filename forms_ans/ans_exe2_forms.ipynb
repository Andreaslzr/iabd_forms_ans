```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dados_setores = pd.DataFrame({
    'setores':[1,2,3,4,5,6,7,8],
    'produtos fabricados': [100,50,15,200,500,1000,375,450],
})
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(dados_setores)
```

```python
sse = kmeans.inertia_
print(sse)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(dados_setores['setores'],dados_setores['produtos fabricados'],c=labels)
plt.title('produtos fabricados por setor')
plt.xlabel('setores')
plt.ylabel('produtos fabricados')
```

```python
dados_setores['grupo'] = labels
grouped_data = dados_setores.groupby('grupo').mean()
print(grouped_data)
```
