```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dados_movimento = pd.DataFrame({
    'dia':[1,2,3,4,5,6,7],
    'numero clientes': [15,20,30,100,350,500,700],
})
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(dados_movimento)
```

```python
sse = kmeans.inertia_
print(sse)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(dados_movimento['dia'],dados_movimento['numero clientes'],c=labels)
plt.title('numero de clientes por dia')
plt.xlabel('dias')
plt.ylabel('numero de clientes')
```

```python
dados_movimento['grupo'] = labels
grouped_data = dados_movimento.groupby('grupo').mean()
print(grouped_data)
```
